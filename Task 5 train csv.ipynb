```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.info
```

```python
df.head()
```

```python
df.describe()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
sns.set(style="whitegrid", palette="muted")
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[numerical_cols].hist(bins=20, figsize=(12,8))
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Survived', y=col, data=df)
    plt.title(f'Boxplot of {col} by Survival')
    plt.show()
    
```

```python
sns.pairplot(df, vars=['Age', 'Fare', 'SibSp', 'Parch'], hue='Survived')
plt.suptitle('Pairplot of Features Colored by Survival', y=1.02)
plt.show()
```

```python
categorical_cols = ['Pclass', 'Sex', 'Embarked']
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Survived', data=df)
    plt.title(f'Survival Count by {col}')
    plt.show()
```

```python
print("""
Observations:
- Higher survival rate among females.
- Passengers in 1st class had higher survival chances.
- Higher fares are associated with higher survival.
- Age has a wide distribution; some very young and old survived.

""")
```
